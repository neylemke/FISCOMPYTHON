## Imports

```python
from altair import *
import pandas as pd
```

## Load dataset

```python
cars = load_dataset('cars')
```

## Default heatmap

```python
Chart(cars).mark_text().encode(
    Color('mean(Miles_per_Gallon):Q', sort='descending'),
    column='Cylinders:O',
    text='mean(Miles_per_Gallon):Q',
    row='Origin'
)
```

## Coloring background

```python
Chart(cars).mark_text(applyColorToBackground=True).encode(
    Row('Origin:O'),
    Column('Cylinders:O'),
    Color('mean(Miles_per_Gallon):Q', sort='descending'),
    Text('mean(Miles_per_Gallon):Q')
)
```

## Customizing colormap
Vega-lite currently only supports colormaps between two colors.

HTML color names can also be used (e.g., `'#d62728'`).

```python
Chart(cars).mark_text(applyColorToBackground=True).encode(
    Row('Origin:O'),
    Column('Cylinders:O'),
    Color('mean(Miles_per_Gallon):Q', sort='descending', scale=Scale(range=['pink', 'red'])),
    Text(' '),
)
```
