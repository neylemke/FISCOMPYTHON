```python
from random import random
list_1 = [random() for _ in range(1000000)]
list_2 = [random() for _ in range(1000000)]
```

```python
out = [x + y for (x, y) in zip(list_1, list_2)]
```

```python
out
```

```python
%timeit [x + y for (x, y) in zip(list_1, list_2)]
```
